## Conditional Selection

```python
# Q. Select all the row labels whose Party is Independent (elections["Party"]=="Independent")

# 1st step: store the result in a boolean  series
elections["Party"]=="Independent"

# 2nd step: pass the series as an input to [ ] selection operator
elections[elections["Party"]=="Independent"]
```

```python
# Q. Select all candidates who won the election with vote percentage >47%
#1.
(elections["Result"]=="win") & (elections["%"]>47)

#2.
elections[(elections["Result"]=="win") & (elections["%"]>47)]
```

```python
# All candidates from Democratic and Republican Parties

parties=["Democratic","Republican"]

#1. Boolean series
elections[elections["Party"].isin(parties)]

#2. query function
elections.query('Party in @parties')
```

```python
# Q. Select all candidates whose names start with "W"
elections[elections["Candidate"].str.startswith("W")]
```

```python
# Starts with more than one character
elections[elections["Candidate"].str.startswith("Wi")]
```
